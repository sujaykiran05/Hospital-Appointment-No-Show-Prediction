```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("HOSPITAL NO SHOW.CSV")
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df['travel_time_min'] = df['travel_time_min'].fillna(df['travel_time_min'].mean())
df['email_reminder'] = df['email_reminder'].fillna(df['email_reminder'].mean())
df['distance_km'] = df['distance_km'].fillna(df['distance_km'].mean())
df['employment_status'] = df['employment_status'].fillna(df['employment_status'].mode()[0])
df['education_level'] = df['education_level'].fillna(df['education_level'].mode()[0])
df.head()
```

```python
df['age'] = df['age'].astype(int)
df['email_reminder'] = df['email_reminder'].astype(int)
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='no_show',data=df)
plt.title('total no show')
plt.show()
```

```python
dept_noshows = df[df['no_show']==1].groupby('department').size()
dept_noshows.plot(kind='bar', color='red')
plt.title('No-Shows by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='gender', hue='no_show', data=df)
plt.title('No-Shows by Gender')
plt.show()
```

```python
sns.countplot(x='sms_reminder', hue='no_show', data=df)
plt.title('SMS Reminder Effect')
plt.show()
```

```python
plt.hist(df[df['no_show']==1]['age'], bins=20, color='red')
plt.title('Age of No-Show Patients')
plt.xlabel('Age')
plt.show()
```

```python
sns.countplot(x='city_type', hue='no_show', data=df)
plt.title('No-Shows by City Type')
plt.show()
```

```python
df.to_csv('cleaned hospital no show.csv', index=False)
```
